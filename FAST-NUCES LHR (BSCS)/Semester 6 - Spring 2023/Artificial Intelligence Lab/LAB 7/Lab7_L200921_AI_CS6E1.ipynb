```python
# Aisha Muhammad Nawaz L20-0921 BSCS 6E1 AI LAB 07 
"""
Tic Tac Toe Player
"""

import copy
import math
import random


X = "X"
O = "O"
D="D"
EMPTY = None


def initial_state():
    """
    Returns starting state of the board.
    """
    return [[EMPTY, EMPTY, EMPTY],
            [EMPTY, EMPTY, EMPTY],
            [EMPTY, EMPTY, EMPTY]]
    
def draw_board(board):
    print("-------------")
    for i in range(3):
        print("| ", end="")
        for j in range(3):
            if board[i][j] == EMPTY:
                print(" ", end="")
            else:
                print(board[i][j], end="")
            print(" | ", end="")
        print()
        print("-------------")


def player(board):
    """
    Returns player who has the next turn on a board.
    """
    count = 0
    for i in board:
        for j in i:
            if j:
                count += 1
    if count % 2 != 0:
        return O
    return X


def actions(board):
    """
    Returns set of all possible actions (i, j) available on the board.
    """
    res = set()
    board_len = len(board)
    for i in range(board_len):
        for j in range(board_len):
            if board[i][j] == EMPTY:
                res.add((i, j))
    return res


def result(board, action):
    """
    Returns the board that results from making move (i, j) on the board.
    """
    curr_player = player(board)
    result_board = copy.deepcopy(board)
    (i, j) = action
    result_board[i][j] = curr_player
    return result_board


def get_horizontal_winner(board):
    # check horizontally
    winner_val = None
    board_len = len(board)
    for i in range(board_len):
        winner_val = board[i][0]
        for j in range(board_len):
            if board[i][j] != winner_val:
                winner_val = None
        if winner_val:
            return winner_val
    return winner_val


def get_vertical_winner(board):
    # check vertically
    winner_val = None
    board_len = len(board)
    for i in range(board_len):
        winner_val = board[0][i]
        for j in range(board_len):
            if board[j][i] != winner_val:
                winner_val = None
        if winner_val:
            return winner_val
    return winner_val


def get_diagonal_winner(board):
    # check diagonally
    winner_val = None
    board_len = len(board)
    winner_val = board[0][0]
    for i in range(board_len):
        if board[i][i] != winner_val:
            winner_val = None
    if winner_val:
        return winner_val

    winner_val = board[0][board_len - 1]
    for i in range(board_len):
        j = board_len - 1 - i
        if board[i][j] != winner_val:
            winner_val = None

    return winner_val


def winner(board):
    """
    Returns the winner of the game, if there is one.
    """
    winner_val = get_horizontal_winner(board) or get_vertical_winner(board) or get_diagonal_winner(board) or None
    return winner_val


def terminal(board):
    """
    Returns True if game is over, False otherwise.
    """
    if winner(board) != None:
        return True

    for i in board:
        for j in i:
            if j == EMPTY:
                return False
    return True

def utility(board):
    """
    Returns 1 if X has won the game, -1 if O has won, 0 otherwise.
    """
    winner_val = winner(board)
    if winner_val == X:
        return 1
    elif winner_val == O:
        return -1
    return 0

#------------------------20L-0921------------------------------------------->
rowNext=0
colNext=0

def evalFunction(state):
  #Calculates the evaluation function value of state 
  #Potential Moves left for O (Score O) - Potential Moves left for X (Score X)
  scoreO,scoreX=0,0
   
  #Checking potential moves left ...

  #Vertically:
  for i in range(3):
      flagX,flagO=1,1
      for j in range(3):
          if state[j][i] != EMPTY or state[j][i]!='X':
              flagX = 0
          elif state[j][i]!= EMPTY or state[j][i]!='O':
              flagO= 0
      if(flagX):
       scoreX=scoreX+1
      if(flagO):
       scoreO=scoreO+1
  
  #Horizontally:
  for i in range(3):
    flagX,flagO=1,1
    for j in range(3):
      if state[i][j] != EMPTY or state[i][j]!='X':
        flagX=0
      elif state[i][j]!=EMPTY or state[i][j]!='O':
        flagO=0
    if(flagX):
     scoreX=scoreX+1
    if(flagO):
     scoreO=scoreO+1

  #Diagonally:
  flagX,flagO=1,1
  for i in range(3):
      if state[i][i] != EMPTY or state[i][i]!='X':
          flagX=0
      elif state[i][i]!=EMPTY or state[i][i]!='O':
          flagO=0
  if(flagX):
    scoreX=scoreX+1
  if(flagO):
    scoreO=scoreO+1
  
  flagX,flagO=1,1
  for i in range(3):
      j = 2 - i
      if state[i][j] != EMPTY or state[i][j]!='X':
          flagX=0
      elif state[i][j]!=EMPTY or state[i][j]!='O':
          flagO=0

  if(flagX):
    scoreX=scoreX+1
  if(flagO):
    scoreO=scoreO+1

  return (scoreO-scoreX)

#Assumption: At best a total of 9 moves can be performed by both players. Hence, the depth is chosen to be 9
def alpha_beta_pruining(state,depth=9,playerTurn='O',r=0,c=0,alpha=float('-inf'),beta=float('inf')):
  global rowNext
  global colNext
  
  if(depth==0 or terminal(state) or utility(state)!=0):
    rowNext=r
    colNext=c
    return evalFunction(state)

  if(playerTurn=='O'): #Maximize, as we want computer to win and computer is player O
    
    maximumValue=float('-inf')
    stateChildren=actions(state)
   
    for kid in stateChildren: #Finding row and coloumn vaues of each kid in state children
      i=0
      for row in kid:
        if(i==0):
         r=row 
        else:
         c=row
        i=i+1 

      valueChild=alpha_beta_pruining(result(state,kid),depth-1,'X',r,c,alpha,beta) #Next turn is of player X
      maximumValue=max(valueChild,maximumValue)
      alpha=max(alpha,maximumValue)   
      if(alpha>=beta): #Prune
       return maximumValue

    #After all children explored    
    return maximumValue
  
  elif(playerTurn=='X'): # Minimize
    minimumValue=float('inf')
    stateChildren=actions(state)
    
    for kid in stateChildren:
      i=0
      for row in kid:
        if(i==0):
         r=row 
        else:
         c=row
        i=i+1
      
      valueChild=alpha_beta_pruining(result(state,kid),depth-1,'O',r,c,alpha,beta)    
      minimumValue=min(valueChild,minimumValue)
      beta=min(beta,minimumValue)
      if(alpha>=beta): #Prune
       return minimumValue
    
    return minimumValue

def play_game(): #Human is X and computer is O
    global rowNext
    global colNext
    board = initial_state()
    draw_board(board)
    while not terminal(board):
        print("Enter the row and column number")
        row, col = map(int, input().split())
        board = result(board, (row, col))
        if not terminal(board): 
            alpha_beta_pruining(board)     #Suggests next action for computer
            action = (rowNext,colNext)     #The rowNext & colNext values were updated when the 'alpha_beta_pruining(board)' function was called
            board = result(board, action)  #Peforms next action of computer i.e 'O' player
            draw_board(board)
    print("Winner is ", winner(board))

print('                     -------------------------------------- Welcome to the Tic-Tac-Toe Game!--------------------------------------\n\nYou will have the first turn as Player X:')
play_game()

```
